## Problem 01

Read in the `hw05_prob_01.csv` CSV file and assign it to the `df01` object.

Visually explore the `df01` object following **ALL** required actions listed previously.

You must add more code cells and markdown cells to complete this problem.

#### 01) - SOLUTION

```python
###
df01 = pd.read_csv('hw05_prob_01.csv')
df01.info()
```

```python
df01.nunique()
```

```python
df01.isna().sum()
```

```python
# bar plot for A B C
fig,ax = plt.subplots(1,3,figsize=(15,4))

sns.countplot(data = df01, x = 'A', ax = ax[0])
sns.countplot(data = df01, x = 'B', ax = ax[1])
sns.countplot(data = df01, x = 'C', ax = ax[2])

plt.show()
```

```python
# HISTOGRAMS for NUMERIC COLUNMS
sns.displot(data=df01, x='x', kind = "hist", kde=True)
plt.show()
```

### Conclusion 1
The categories of the NON_NUMERIC columns : column A has similar counts, but column B and C have different counts.

The histograms of the NUMERIC columns don't look like bell curves.

```python
## DODGED BAR PLOT between 2 NON-NUMERIC COLUMNS
sns.catplot(data = df01, x = 'A', hue = 'B', kind = "count")
plt.show()
```

```python
## HEAT MAP between 2 NON-NUMERIC COLUMNS
fig,ax = plt.subplots()

sns.heatmap(pd.crosstab(df01['A'], df01['B']), ax=ax,
            annot = True,
            annot_kws = {'size': 20},
            fmt = 'd',
            cbar = False)

plt.show()
```

### Conclusion 2
No, there are no combinations between the 2 variables NOT present in the data.

HEAT MAP is easier to identify combinations that were NOT present.

```python
## Conditional KDE PLOT on NUMERIC COLUMN, conditioned on NON-NUMERIC COLUMN

sns.displot(data = df01, x = 'x', hue = 'A', kind = "kde",
            common_norm = False)

plt.show()
```

```python
## Boxplot on NUMERIC COLUMN
sns.catplot(data = df01, x = 'A', y = 'x', kind = "box",
            showmeans = True,
            meanprops={'marker':'o','markerfacecolor':'white','markeredgecolor':'black'})
                      
plt.show()
```

```python
## POINT PLOT on NUMERIC COLUMN
sns.catplot(data = df01, x = 'A', y = 'x', kind = "point",
            join = False)

plt.show()
```

### Conclusion 3
1. Yes, they are different.
2. They have similar median
3. They have different variation.
4. Yes, I am confident that the MEAN value CHANGES.
